# TP Matrices 2 : manipulation d'images

```python
from scipy import misc
m = misc.face()
r, c = len(m), len(m[0])
```

```python
f(x) = mean(x)
m = matrix([[int(mean(m[i,j])) for j in range(c)] for i in range(r)])
```

```python
matrix_plot(m, cmap='gray')
```

```python
mp = matrix([[m[2*i,2*j] for j in range(c/2)] for i in range(r/2)])
```

```python
matrix_plot(mp, cmap='gray')
```

```python
mn = matrix([[256 - m[i,j] for j in range(c)] for i in range(r)])
```

```python
matrix_plot(mn, cmap='gray')
```

```python
def mat(m,f, nr = m.nrows(), nc = m.ncols()):
    return matrix_plot(matrix([[f(i,j) for j in range(nc)] for i in range(nr)]), cmap='gray')
```

```python
mat(m,lambda i,j : 256-m[i,j])
```

```python
mat(m,lambda i,j : m[j,i], m.ncols(), m.nrows())
```
